```python
import numpy as np

BASE_NUM=5

Orders = np.zeros((BASE_NUM, BASE_NUM), dtype=np.float64)
print(Orders)

a1 = np.arange(0,BASE_NUM) # index [0]
a2 = np.random.uniform(1.0,8.0,BASE_NUM) # price [1]
a3 = np.random.uniform(0.5,4.0,BASE_NUM) # amount [2]
a4 = np.zeros((BASE_NUM,)) # contract amount [3]
a5 = np.stack([a1, a2, a3, a4], 1)

Node_Sellable_Rate = sorted(a5, key=lambda x: x[1])

b1 = np.arange(0,BASE_NUM)
b2 = np.random.uniform(1.0,8.0,BASE_NUM)
b3 = np.random.uniform(0.5,4.0,BASE_NUM)
b4 = np.zeros((BASE_NUM,))
b5 = np.stack([b1, b2, b3, b4], 1)

Node_Buyable_Rate = sorted(b5, reverse=True, key=lambda x: x[1])

print("=====origin=========================")
print("index, price, amount, result")
print("===============Seller===============")
print(*Node_Sellable_Rate, sep='\n')

print("===============Buyer===============")
print(*Node_Buyable_Rate, sep='\n')
print("====================================")

remain = 0
remain_b = 0
isFinal = False

for sell in Node_Sellable_Rate:
    for buy in Node_Buyable_Rate:
        if (buy[2] == 0):
            continue

        if (sell[1] <= buy[1]): # compare price
            print(round(sell[1],1), "<=", round(buy[1],1))
            if (remain == 0):
                remain = sell[2] - buy[2]
                print(remain, "=", round(sell[2],3), "-", round(buy[2],3))
            elif (remain > 0):
                remain_b = remain
                remain = remain - buy[2]
                print(remain, "=", round(remain_b,3), "-", round(buy[2],3))
            else:
                remain_b = remain
                remain = sell[2] + remain
                print(remain, "=", round(sell[2],3), "-", round(remain_b,3))
            
            if (remain >= 0):
                if remain_b >= 0:
                  print("  p1: ", int(sell[0]), "->", int(buy[0]), ": ", buy[2])
                  Orders[int(sell[0]), int(buy[0])] = buy[2]
                else:
                  print("  p2: ", int(sell[0]), "->", int(buy[0]), ": ", remain_b)
                  Orders[int(sell[0]), int(buy[0])] = - remain_b
                buy[3] = buy[2]
                buy[2] = 0
            else:
                if remain_b > 0:
                  print("  p3: ", int(sell[0]), "->", int(buy[0]), ": ", remain_b)
                  Orders[int(sell[0]), int(buy[0])] = remain_b
                else:
                  print("  p4: ", int(sell[0]), "->", int(buy[0]), ": ", sell[2])
                  Orders[int(sell[0]), int(buy[0])] = sell[2]
                sell[3] = sell[2]
                sell[2] = 0
                break
        else:
          if (not isFinal):
              isFinal = True
              if (remain >= 0):
                  sell[3] = sell[2] - remain
                  print("set sell result of ", int(sell[0]), "=", sell[3])
                  break
              else:
                  buy[3] = buy[2] + remain
                  print("set buy result of ", int(buy[0]), "=", buy[3])
                  break
              

print("====================================")
print("")
print("")
print("=====result=========================")
print("index, price, amount, result")
print("===============Seller===============")
print(*Node_Sellable_Rate, sep='\n')

print("===============Buyer===============")
print(*Node_Buyable_Rate, sep='\n')

print("===============Check Sum============")
print(np.array(Node_Buyable_Rate)[:,2].sum())
print(np.array(Node_Buyable_Rate)[:,2].sum())

print("====================================")
print(Orders)
print(Orders.sum(axis = 1))
```

```python
import numpy as np

BASE_NUM=5

Orders = np.zeros((BASE_NUM, BASE_NUM), dtype=np.float64)

for time in range(1, 5):
    np.random.seed(time)
    a1 = np.arange(0,BASE_NUM) # index [0]
    a2 = np.random.uniform(1.0,8.0,BASE_NUM) # price [1]
    a3 = np.random.uniform(0.5,4.0,BASE_NUM) # amount [2]
    a4 = np.zeros((BASE_NUM,)) # contract amount [3]
    a5 = np.stack([a1, a2, a3, a4], 1)

    Node_Sellable_Rate = sorted(a5, key=lambda x: x[1])

    b1 = np.arange(0,BASE_NUM)
    b2 = np.random.uniform(1.0,8.0,BASE_NUM)
    b3 = np.random.uniform(0.5,4.0,BASE_NUM)
    b4 = np.zeros((BASE_NUM,))
    b5 = np.stack([b1, b2, b3, b4], 1)

    Node_Buyable_Rate = sorted(b5, reverse=True, key=lambda x: x[1])

    remain = 0
    remain_b = 0
    isFinal = False

    for sell in Node_Sellable_Rate:
        for buy in Node_Buyable_Rate:
            if (buy[2] == 0):
                continue

            if (sell[1] <= buy[1]): # compare price
                if (remain == 0):
                    remain = sell[2] - buy[2]
                elif (remain > 0):
                    remain_b = remain
                    remain = remain - buy[2]
                else:
                    remain_b = remain
                    remain = sell[2] + remain
                
                if (remain >= 0):
                    if remain_b >= 0:
                      Orders[int(sell[0]), int(buy[0])] = Orders[int(sell[0]), int(buy[0])] + buy[2]
                    else:
                      Orders[int(sell[0]), int(buy[0])] = Orders[int(sell[0]), int(buy[0])] - remain_b
                    buy[3] = buy[2]
                    buy[2] = 0
                else:
                    if remain_b > 0:
                      Orders[int(sell[0]), int(buy[0])] = Orders[int(sell[0]), int(buy[0])] + remain_b
                    else:
                      Orders[int(sell[0]), int(buy[0])] = Orders[int(sell[0]), int(buy[0])] + sell[2]
                    sell[3] = sell[2]
                    sell[2] = 0
                    break
            else:
              if (not isFinal):
                  isFinal = True
                  if (remain >= 0):
                      sell[3] = sell[2] - remain
                      break
                  else:
                      buy[3] = buy[2] + remain
                      break
print(Orders)
print(Orders.T)
print(Orders.T.sum(axis = 0))
print(Orders.T/Orders.T.sum(axis = 0))
```
