```python
import numpy as np

BASE_NUM = 20

a1 = np.arange(0,BASE_NUM) # index [0]
a2 = np.random.uniform(1.0,8.0,BASE_NUM) # price [1]
a3 = np.random.uniform(0.5,4.0,BASE_NUM) # amount [2]
a4 = np.zeros((BASE_NUM,)) # contract amount [3]
a5 = np.stack([a1, a2, a3, a4], 1)

Node_Sellable_Rate = sorted(a5, key=lambda x: x[1])

b1 = np.arange(0,BASE_NUM)
b2 = np.random.uniform(1.0,8.0,BASE_NUM)
b3 = np.random.uniform(0.5,4.0,BASE_NUM)
b4 = np.zeros((BASE_NUM,))
b5 = np.stack([b1, b2, b3, b4], 1)

Node_Buyable_Rate = sorted(b5, reverse=True, key=lambda x: x[1])

print("=====origin=========================")
print("index, price, amount, result")
print("===============Seller===============")
print(*Node_Sellable_Rate, sep='\n')

print("===============Buyer===============")
print(*Node_Buyable_Rate, sep='\n')
print("====================================")

remain = 0

for sell in Node_Sellable_Rate:
    for buy in Node_Buyable_Rate:
        if (buy[2] == 0):
            continue

        if (sell[1] <= buy[1]): # compare price
            print(sell[1], "<=", buy[1])
            if (remain == 0):
                remain = sell[2] - buy[2]
                print(remain, "=", sell[2], "-", buy[2])
            elif (remain > 0):
                remain_b = remain
                remain = remain - buy[2]
                print(remain, "=", remain_b, "-", buy[2])
            else:
                remain_b = remain
                remain = sell[2] + remain
                print(remain, "=", sell[2], "-", remain_b)
            
            if (remain >= 0):
                buy[3] = buy[2]
                buy[2] = 0
            else:
                sell[3] = sell[2]
                sell[2] = 0
                break
        else:
          if (remain >= 0):
              sell[3] = sell[2] - remain
          else:
              buy[3] = buy[2] + remain

print("====================================")
print("")
print("")
print("=====result=========================")
print("index, price, amount, result")
print("===============Seller===============")
print(*Node_Sellable_Rate, sep='\n')

print("===============Buyer===============")
print(*Node_Buyable_Rate, sep='\n')
```
